```python
#import packages
import numpy as np
# Import de funcion generica
from utils_distribuciones import generar_distribucion
```

```python
media = 0
sigma = 1

dist_normal = np.random.normal(media, sigma, 1000)
```

```python
# Uso de nuestra funcion generica
generar_distribucion(dist_normal, 'Distribucion normal')
```

```python
dist_sesgada = np.random.exponential(1, 1000)

# Uso de nuestra funcion generica
generar_distribucion(dist_sesgada, 'Distribucion Sesgada a la Derecha')
```

```python
dist_sesgada_izq = -np.random.exponential(1, 1000)

# Uso de nuestra funcion generica
generar_distribucion(dist_sesgada_izq, 'Distribucion Sesgada a la Izquierda')
```

## Distribucion Bimodal

```python
dist_bimodal = np.concatenate([
                               np.random.normal(-2, 1, 500), 
                               np.random.normal(3, 1, 500)
                              ])
generar_distribucion(dist_bimodal,'Distribucion Bimodal')
```
